```python
alpha, eps, threshold = 1, 5, .5
scores = [i*0.05 for i in range(-10, 20)]

def cos_sim_score_with_threshold(score, eps, alpha, threshold):
    
    '''        
        Gets the score identifies to increase or decrease it and returns new update score.
        
        Arguments:
        
        score - score within range of 0~1;
        eps - epsilon value, hyperparameter;
        alpha - alpha value, hyperparameter;
        threshold - value to classify the score as positive or negative.
    '''   
    
    if score >= threshold:
        print("Classified as positive, boosting the score!")
        return (score + eps) / (eps + alpha)
    elif score < threshold:
        print("Classified as negative, decreasing the score!")
        return (score + (alpha / eps)) / (2*eps)

def cos_sim_score_booster(score, eps, alpha, mode):
    
    '''        
        Gets the score and mode and returns new update score based on the mode.
        
        Arguments:
        
        score - score within range of 0~1;
        eps - epsilon value, hyperparameter;
        alpha - alpha value, hyperparameter;
        mode - mode to increase or decrease the score.
    '''   
    
    if mode == "for_pos":
        return (score + eps) / (eps + alpha)
    elif mode == "for_neg":
        return (score + (alpha / eps)) / (2*eps)

funcs = {"Threshold_based booster": cos_sim_score_with_threshold, "Manual booster": cos_sim_score_booster}

for func_name, func in funcs.items():
    for i, score in enumerate(scores):
        print(f"\nReal score: {score:.3f}")
        if func_name == "Threshold_based booster":
            print(f"\nBoosting with {func_name}...")
            print(f"New score: {func(score, eps, alpha, threshold):.3f}")
        elif func_name == "Manual booster":
            print(f"\nBoosting with {func_name}...")
            print(f"New positive score: {func(score, eps, alpha, 'for_pos'):.3f}")
            print(f"New negative score: {func(score, eps, alpha, 'for_neg'):.3f}")
```
